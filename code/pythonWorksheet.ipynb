```python
def missingValCheck(df):
    count = df.isnull().sum()
    return count
```

```python
missingValCheck(df)
```

```python
def medianFiller(df, colString):
    df[colString].fillna(df[colString].median(),inplace=True)
    return df
```

```python
medianFiller(df, "age")
```

```python
#2 Duplicate Checker

def duplicateCheck(df, remove=False):
    original_count = df.duplicated().sum()
    if remove==True:
        cleanDF = df.drop_duplicates()
    return original_count, cleanDF
```

```python
count, df = duplicateCheck(df, remove=True)
df.head()
```

```python
def outlierChecker(df, colString, viewOutliers=False):
    q1 = df[colString].quantile(0.25)
    q3 = df[colString].quantile(0.75)
    IQR = q3-q1
    
    if viewOutliers == True:
        answer = df[(df[colString] < q1-(1.5*IQR)) | (df[colString] > q3+(1.5*IQR))]
    elif viewOutliers == False:
        answer = df[~((df[colString] < q1-(1.5*IQR)) | (df[colString] > q3+(1.5*IQR)))]
    return answer
```

```python
outlierChecker(df, "age",viewOutliers=False)
```

```python
#5 Create family size based on "SibSp" and "Parch".

df["FamilySize"] = df['sibsp'] + df['parch'] + 1
df.head()
```

```python
#6 Convert embarked column to numerical one-hot encoding
from sklearn.preprocessing import LabelEncoder

def numCode(df, colString):
    encoder = LabelEncoder()
    df[colString] = encoder.fit_transform(df[colString])
    return df
```

```python
numCode(df, "embarked")
```

```python
#8 Clean name column and make everything lower case - NOT FINISHED

import re

def nameCleaner(df, colString):
    df[colString] = df[colString].apply(lambda x: re.sub(r'mr\.mrs\.|dr\.','', x))
    df[colString] = df[colString].str.lower()
    return df
```

```python
df1 = nameCleaner(df, "name")
df1.head()
```
